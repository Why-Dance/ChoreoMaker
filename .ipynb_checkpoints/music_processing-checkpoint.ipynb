# Youtubeì—ì„œ ë…¸ë˜ ì œëª©ìœ¼ë¡œ ê²€ìƒ‰,  ê°€ì¥ ì²«ë²ˆì§¸ì˜ ì˜ìƒ ê°€ì ¸ì˜¤ê¸°

```python
download_base_url = 'https://www.youtube.com/watch?v='

def youtube_search(options,list_name):
    youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,
    developerKey=DEVELOPER_KEY)

    search_response = youtube.search().list(
    q=options["q"],
    part="id,snippet",
    maxResults=options["maxResults"]
    ).execute()
    
    for search_result in search_response.get("items",[]):
        print("===========================")
        print(type(search_result))
        print(search_result["id"])
        print(type(search_result["id"]))
        print(search_result["snippet"])
        print(type(search_result["snippet"]))
        print("===========================")
        
        
    video = {}
    
    for search_result in search_response.get("items", []):
        if search_result["id"]["kind"] == "youtube#video":
            video = {"title": search_result["snippet"]["title"],"url":download_base_url+search_result["id"]["videoId"]}
            list_name.append(video)

    return list_name #return results with list
```

# ffmpegë¥¼ í†µí•´ .wavë¡œ ë³€í™˜

```python
f = []
for (dirpath, dirnames, filenames) in walk(moriginal_path):
    f.extend(filenames)
    break

# Convert downloaded music into wav format
musList = []
for vid in f:
    m = vid.split(".")[0]
    musList.append(m)
    print(m)
    print(musList)
    os.system("cd {p} && ffmpeg -i \"{f0}\\{f1}\" \"{f2}\\{f3}.wav\"".format(p=proj_path,f0=moriginal_path,f1=vid, f2=mwav_path, f3=m))
```

# beat ì°íŒ ì´ˆë¥¼ text fileë¡œ ë§Œë“¤ê¸°

```python
#dictionaries
beat_dic = {}

for fname in musList:
    cmd =  "python DBNBeatTracker single -o \"{v2}\\{v3}.txt\" \"{v4}\\{v5}.wav\"".format(v1=madmom_b_path, v2=mbeat_path,v3=fname,v4=mwav_path,v5=fname)
    os.system('cd {v1} && {c}'.format(v1=madmom_b_path, c=cmd))
    
    beat_sec = []
    with open(mbeat_path+"\\"+fname+".txt") as f:
        for line in f:
            beat_sec.append(float(line))
            
    beat_dic[fname] = beat_sec
```
